# 0. 자기회귀과정 ( Autoregressive Process : AR)

* 현재의 관측값이 과거의 관측값들의 함수형태로 표현

$$ Z_t = f(Z_{t-1},Z_{t-2}\dots) + \varepsilon_t$$

* $Z_t-\mu = \phi_1(Z_{t-1}-\mu)+\phi_2(Z_{t-2}-\mu)+\dots \varepsilon_t$

* 이를 후진연산자 $B^j\,(Z_{t}-\mu) =Z_{t-j}-\mu$ 을 이용하면

$$\begin {eqnarray*} Z_t - \mu &=& (Z_t-\mu)\sum_{j=1}^n \phi_j B^j +\varepsilon_t\end{eqnarray*}$$ 

* 따라서 $ \left(1-\sum_{i=1}^n \phi_j B^j\right) (Z_t-\mu)=\varepsilon_t$ 임이 자명하고 이를 다시

* $\phi(B) (Z_t-\mu) = \varepsilon_t$라고 쓸 수 있다. $\phi(B)$ is AR operator

## 1. AR(1)

$$Z_t -\mu = \phi (Z_{t-1}-\mu)+\varepsilon_t,\quad \varepsilon_t \sim WN(0,\sigma^2)$$

* 이는 $(1-\phi_1B)(Z_t-\mu)=\varepsilon_t$로 표현할 수 있음

* 또한  $Z_t = \mu + \sum_{j=0}^\infty \phi^j\varepsilon_{t-j}$ 과 같다.

* $|\phi|<1$ 이라는 조건은 AR(1)이 정상성을 만족하기 위한 조건이다.

* $E(Z_t) = \mu + \phi^t(E(Z_0)-\mu)$  $\quad if,\,\,|\phi|<1\quad E(Z_t)=\mu$

* $\gamma_0=Var(Z_t) = \phi^2Var(Z_{t-1})+\sigma^2 = \sigma^2\sum_{j=0}^\infty\phi^{2j}$  $\quad if,\,\,|\phi|<1\quad Var(Z_t)=\frac{\sigma^2}{1-\phi^2}$

* $\gamma_h=Cov (Z_t, Z_{t+h}) = \phi^h\frac{\sigma^2}{1-\phi^2} = \phi^2Var(Z_t)\quad h \geq1$

* $\begin{eqnarray*}\rho_h = \frac{\gamma_h}{\gamma_0} = \phi^h, \quad h\geq1 \end{eqnarray*}$

* 정상일 조건 $|\phi| < 1$이므로 $\rho_h$는 시차 $h$가 커짐에 따라 지수적으로 감소한다.

* 부분자기상관계수

$\begin{eqnarray*}PACF = \phi_{kk} = \begin{cases}\rho_1, \quad k=1 \\ \\
                                       0,\quad k\geq 2  \end{cases} \end{eqnarray*}$

## 2. AR(2)

* AR(1)이 $Z_t$를 고려할 때 이전 시점까지 고려했다면 AR(2)는 2-시차 만큼 고려한다.

$$Z_t - \mu = \phi_1(Z_{t-1}-\mu)+\phi_2(Z_t-\mu)+\varepsilon_t,\quad \varepsilon_t \sim WN(0,\sigma^2)$$

* $\mu$에 대하서 정리하면 $\quad Z_t = \delta +\phi_1Z_{t-1}+\phi_2Z_{t-2}+\varepsilon_t,\quad \delta = (1-\phi_1-\phi_2)\mu$

* 후진연산자를 이용하면 $(1-\phi_1B -\phi_2B^2)(Z_t-\mu) =\varepsilon_t \quad\to \quad \phi(B)(Z_t-\mu)=\varepsilon_t$

* 정상성의 조건은 다음과 같다. => $\phi(B)$의 근의 절댓값이 1보다 클 때 즉 아래와 같다.

$\qquad \begin{cases} \phi_1+\phi_2 <1 \\ \\
               \phi_2-\phi_1 <1 \\ \\ 
               -1<\phi_2<1\end{cases}$

* 정상성을 만족할 때는 다음을 만족한다.(이 부분 다시)

  * $Z_t = \mu + \sum_{i=1}^2\sum_{j=0}^\infty \phi_i^j\varepsilon_{t-j}$
  * $E(Z_t) = \mu$
  * $\begin{eqnarray*}Var(Z_t) = \sigma^2 \sum_{j=0}^\infty \phi_j^2 = \frac {1-\phi_2}{1+\phi_2}\times \frac {\sigma^2}{(1-\phi_2)^2-\phi_1^2}\end{eqnarray*}$

* $\rho_k = \begin{cases} \frac{\phi_1}{1-\phi_2} \quad k=1\\ \\ 
                           \frac{\phi_1^2+\phi_2-\phi_2^2}{1-\phi_2} \quad k=2 \\ \\
                           \phi_1\rho_{k-1}+\phi_2\rho_{k-2},\quad k\geq3 \end{cases}$

* $\phi_{kk} = \begin{cases} \rho_1 ,\quad k=1\\ \\ 
                            \phi_2 =\frac{\rho_2-\rho_1^2}{1-\rho_1^2},\quad k=2 \\ \\
                            0,\quad k\geq 3\end{cases}$

* AR(2) 과정에서는 $\phi(B)=0$의 두 근이 모두 실수 일때는 ACF가 지수적으로 감소하고, 두 근이 모두 복소수 일 때는 점차 소멸하는 싸인 함수의 형태를 보인다.

* 또한 이론적으로 PACF는 시차 2이후부터는 0이되는 절단 형태를 보인다.

## 3. AR(P) = 일반화

* 정상성의 조건은 $\phi (B)=0$의 근의 절대값이 1보다 클 때이다.

* 정상성을 만족할 때 아래가 성립한다.

$Z_t = \mu + \sum_{j=0}^\infty \psi_j\varepsilon_{t-j}$

$E(Z_t)=\mu,Var(Z_t) = \sigma^2 \sum_{j=0}^\infty \psi_j^2$

$\rho_k = \phi_1\rho_{k-1}+\dots \phi_p \rho_{k-p}, \quad k\geq 1$

$\phi_{kk} = 0, \quad k>p$

# 1. 이동평균과정 (Moving Average Process)

* 현재의 관측값이 과거의 오차값들의 함수로 표현

## 1. MA(q) Process

$$Z_t - \mu = \varepsilon_t -\theta_1\varepsilon_{t-1}-\dots \theta_q\varepsilon_{t-q},\quad \varepsilon_t \sim WN(0,\sigma^2)$$

$$ Z_t - \mu =(1-\theta_1B -\theta_qB^q)\varepsilon_t \quad=>\quad (Z_t-\mu) = \Theta(B)\varepsilon_t$$

* $\Theta(B) : $ MA Operator

* 선형시계열(Stationary linear process)의 특별한 경우 : 언제나 정상시계열이다.

* $Z_t = \mu +\sum_{j=0}^{\infty} \psi_j \varepsilon_{t-j}\quad \sum_{j=0}^\infty{\psi_j^2}<\infty$

* $\psi_0=1,\psi_1=-\theta_1\dots \psi_q=-\theta_q, \quad k \leq q$

* $\psi_k = 0,\quad k>q$ 

* $E(Z_t) = \mu,\quad Var(Z_t)= (1+\theta_1 ^2+\dots \theta_q^2)\,\sigma^2$

* ACF 값을 보고 MA의 p값을 알 수 있다.

* $\rho_h=0 ,\quad h>q$

* $\phi_{kk}$ 절대값이 지수적으로 감소

* 가역성 조건
    * $\Theta(B) = 1-\theta_1B-\dots-\theta_qB^q=0$의 근의 절대값이 1보다 크다.

## 2. MA(1) Process

$$Z_t -\mu = \varepsilon_t - \theta_1\varepsilon_{t-1}\quad \varepsilon_t \sim WN(0,\sigma^2)$$

* $Z_t-\mu = (1-\theta B)\varepsilon_t = \Theta(B)\varepsilon_t$

* 언제나 정상확률과정

* $\varepsilon_t= \sum_{j=0}^\infty \theta_1^jZ_{t-j}$

* $E(Z_t) = \mu,\quad Var(Z_t) = (1+\theta_1^2)\sigma^2$

* $Cov(Z_t,Z_{t-1}) =\theta\sigma^2\to \gamma_1 = -\theta \sigma^2$

* $\begin{eqnarray*} \rho_k = \begin{cases} -\frac{\theta_1}{1+\theta_1^2} = \frac{\gamma_1}{\gamma_0} \quad k=1 \\ \\ 
                                             0,\quad k\geq 2   \\\end{cases}\end{eqnarray*}$

* $\phi_{kk} = \begin{eqnarray*}\frac {-\theta^k(1-\theta^2)}{1-\theta^{2(k+1)}} \end{eqnarray*}\quad k\geq1$

## 3. 가역성 (invertibility) - MA(1)

* 어렵게 생각하지말고 AR process의 정상성과 동일하다.

* 정의 : 어떤 ACF와 PACF를 가지는 MA process는 하나만 존재한다.

* $|\theta|<1$

* $\Theta(B) = 1-\theta B=0$ 의근 $B= 1/\theta$의 절대값이 1보다 크다.

* 하나의 ACF에 하나의 모형이 대응이 되야하기 때문에 가역성을 부과한다.

* 관측 불가능한 오차항을 과거의 관측값들로 표현 가능 - > 잔차를 쉽게 구할 수 있음

## 4. MA(2)

$$Z_t - \mu = \varepsilon_{t}-\theta_1\varepsilon_{t-1}-\theta_2\varepsilon_{t-2},\quad \varepsilon_t \sim WN(0,\sigma^2) $$

*  $Z_t -\mu = (1-\theta_1B-\theta_2B^2)\varepsilon_t =\Theta(B)\varepsilon_t$

* 가역성 조건
    * $\Theta(B) = 1-\theta_1B-\theta_2B^2=0$의 근의 절댓값이 1보다 크다
    * $\theta_1+\theta_2 <1, \theta_2-\theta_1<1, -1<\theta_2<1$

# 2. 자기회귀과정과 이동평균과정의 쌍대성(duality)

* $AR(p) \to MA(\infty)$ 또한 $MA(q) \to AR(\infty)$ 로 표현가능

앞에서 $AR(p)\sim Z_t = \mu + \sum_{j=0}^\infty \psi_j \varepsilon_{t-j}$ 이므로 위가 성립한다.

* $Z_t$는 오차항들의 합이 표현가능하고, $\varepsilon_t$ 는 관측값들의 합으로 표현 가능하기 때문에 쌍대성의 원리를 적용할 수 있는 것이다.

* 또한 $\varepsilon_t$를 과거의 관측값들의 함수로 표현이 가능다는 것은 가역이 가능하다는 뜻이다.

* 즉, 정상성($AR(p)$)을 만족하고 가역성($MA(q)$)을 만족하면 무한차원의 모형을 유한차원의 유형으로 변경할 수 있다.

# 3. ARMA (p,q) : 자기회귀이동평균과정

$$Z_t -\mu = \phi_1(Z_{t-1}-\mu)\dots \phi_p(Z_{t-p}-\mu)+\varepsilon_t -\theta_1\varepsilon_{t-1}-\dots \theta_q\varepsilon_{t-q}$$

* 즉 앞서말한 쌍대성의 원리를 이용해 무한차원의 모형을 유한차원의 모형으로 표현하자는 원리이다.

* 무한개의 모수를 구해야 하는것을 $p+q$개의 모수만을 고려하자는 것이 $ARMA(p,q)$모형이 필요한 이유이다.

* $\varepsilon_t \sim WN(0,\sigma^2)$

* $(1-\phi_1 B-\dots -\phi_pB^q)(Z_t-\mu)=(1-\theta_1B-\dots -\theta_qB^q)\varepsilon_t$

* 특성함수
    * AR polynomial : $\Phi(B) = 1-\phi_1B\dots \phi_pB^p$
    * MA polynomial : $\Theta(B) = 1-\theta_1B\dots \theta_pB^p$
    

* $\Phi(B),\Theta(B)$는 공통인수를 가지지 않으며 정상성과 가역성을 만족해야한다.

## ARMA(1,1)

* $\begin{eqnarray*}E(Z_t) = \mu,\quad Var(Z_t) = \frac{(1+\theta^2-2\phi\theta)}{1-\phi^2}\sigma^2 \end{eqnarray*}$

* $\begin{eqnarray*}\rho_k =\begin{cases} \frac {(\phi-\theta)(1-\phi\theta)}{1+\theta^2-2\phi\theta} \quad k=1 \\ \\ 
                          \phi^{k-1}\rho_1 \quad k\geq 2  \end{cases} \end{eqnarray*} $

* AR(1) 과정의 경우는 감소하는 형태가 $\rho_1=1$부터 시작하나 ARMA(1,1)에서는 $\rho_2$부터 지수적으로 감소한다.
