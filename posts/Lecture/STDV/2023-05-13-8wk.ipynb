# groupyby

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/flights.csv')
df.head()
```

## 특정변수기준으로 나누고 평균을 구하자.

`-` 방법 1

```python
df.groupby("AIRLINE").aggregate({"ARR_DELAY" : "mean"})
```

`-` 방법 2

```python
df.groupby("AIRLINE").aggregate({"ARR_DELAY":np.mean})
```

`-` 방법 3

```python
df.groupby("AIRLINE")["ARR_DELAY"].agg("mean")
```

`-` 방법 4

```python
df.groupby("AIRLINE")["ARR_DELAY"].agg(np.mean)
```

`-` 방법 5

```python
df.groupby("AIRLINE")["ARR_DELAY"].mean()
```

`-` 방법 2와 방법 4는 사용자정의 함수를 쓸 수 있다는 장점이 있음.

```python
def f(x) : return -np.mean(x)
```

```python
df.groupby("AIRLINE").agg({"ARR_DELAY" : f})
```

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY':lambda x: -np.mean(x)})
```

`-` 입력이 여러개인 사용자 정의 함수도 사용가능함 

```python
def f(x,y): return np.mean(x)**y 
```

```python
df.groupby(by='AIRLINE')['ARR_DELAY'].agg(f,2)
```

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY': lambda x: f(x,2)})
```

## AIRLINE,WEEKDAY  $\to$  {CANCELLED: sum}

```python
df.groupby(["AIRLINE","WEEKDAY"]).agg({"CANCELLED":sum})
```

## AIRLINE,WEEKDAY  $\to$ {CANCELLED: sum, mean} , {DIVERTED: sum, mean} 

```python
df.groupby(["AIRLINE","WEEKDAY"]).agg({"CANCELLED" : [sum,np.mean],
                                      "DIVERTED" : [sum,np.mean],
                                      })
```

## AIRLINE,WEEKDAY  $\to$ {CANCELLED: sum, mean, size} , {AIR_TIME: mean,var}

```python
df.groupby(["AIRLINE","WEEKDAY"]).agg({ "CANCELLED" : [sum,np.mean,len],
                                         "AIR_TIME" : [np.mean,np.var]})
```

## cut 함수

`-` 목표 :  DIST를 적당한 구간으로 나누어 카테고리화 하고 그것을 바탕으로 groupby를 수행하자.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.DIST.hist()
```

```python
df.DIST.describe()
```

`-` 구간을 아래와 같이 설정한다.

```python
bins = [-np.inf, 400,700,1200,np.inf]
```

```python
bins
```

```python
cuts= pd.cut(df.DIST,bins=bins).to_list()
```

```python
df["cuts"] = cuts
```

```python
df.groupby(["cuts","AIRLINE"]).agg({"DIVERTED":sum})
```

```python
df.groupby(['AIRLINE']).agg({'AIRLINE':len})
```

`-` cuts을 이용하여 추가그룹핑을 하면 조금 다른 특징들을 데이터에서 발견할 수 있다. 
- AA항공사와 DL항공사는 모두 비슷한 우회횟수를 가지고 있음. 
- AA항공사는 700회이상의 구간에서 우회를 많이하고 DL항공사는 400~700사이에서 우회를 많이 한다. (패턴이 다름) 

`-` 구간이름에 label을 붙이는 방법 

```python
bins
```

```python
cuts2=pd.cut(df.DIST,bins=bins,labels=['Q1','Q2','Q3','Q4'])
cuts2
```

```python
df.groupby(by=[cuts2,'AIRLINE']).agg({'DIVERTED':sum})
```

```python
df.groupby(cuts2).agg({'DIVERTED':len})
```

# Barplot + 해들리위컴의 그래프레이어

```python
g=['A']*100+['B']*200 
y=list(np.random.randn(100)*2+2)+list(np.random.randn(200)+3)
df=pd.DataFrame({'g':g,'y':y})
df
```

`-` geom_bar : 특정 컬럼값을 디폴트로 카운트함

```python
(
ggplot(df) +
    geom_bar(aes(x="g",fill="g"))
)
```

`-` 위 코드는 다음 코드를 디폴트로 한것이다.

```python
(
   ggplot(df.groupby(by='g').count().reset_index()) +
    geom_bar(aes(x="g",y="y",fill="g"),stat="identity")
)
```

```python
td
```

# 심슨의 역설

`-` 아래와 같은 대학합격률 데이터가 있다고 하자.

## 시각화 1

```python
DEP=(['A1']*2+['A2']*2+['B1']*2+['B2']*2)*2 
GEN=['M']*8+['F']*8
STATE=['PASS','FAIL']*8
COUNT=[1,9,2,8,80,20,85,15,5,5,5,5,9,1,9,1]
```

```python
df=pd.DataFrame({'DEP':DEP,'STATE':STATE,'GEN':GEN,'COUNT':COUNT})
df
```

```python
df1=df.groupby(['GEN','STATE']).agg({'COUNT':np.sum}).reset_index()
g_sum =df1.groupby("GEN")["COUNT"].sum().to_list()
g_sum =list(np.repeat(g_sum,2))
df1["prop"] = df1["COUNT"]/g_sum
df1
```

```python
(
    ggplot(df1.query("STATE =='PASS'")) +
    geom_bar(aes(x="GEN",y="prop",fill="GEN"),stat="identity")
)
```

`-` 남학생의 합격률이 더 높아 보인다. $\to$ 남녀차별인가?

## 시각화 2: 학과별 합격률

```python
td =df.groupby(["DEP","GEN"]).agg({"COUNT":sum})\
          .reset_index().\
            rename(columns={"COUNT": "SUM"}).merge(df)
```

```python
td["prop"] = td.COUNT/td.SUM
```

```python
td
```

```python
(
ggplot(td.query("STATE=='PASS'")) +
    geom_bar(aes(x="GEN",y="prop",fill="GEN"),stat="identity") +
    facet_wrap("DEP")
)
```

* 즉 전체적으로 보았을 때는 남학생의 합격률이 더 높은 것 같지만, 학과별로 변수를 나누어서 살펴본 결과 여학생의 합격률이 더 높았다는 것이다.
