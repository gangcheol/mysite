# 최소 평균제곱오차 예측

* 평균제곱오차 : MSE

$$ MSE = E\,[(Y-\hat{Y})^2]$$

* 최소 평균제곱오차(MMSE) 예측 : LSE를 이용하여 모수를 추정후 MSE를 구하는 것 그냥 여태가지 계속해왔던 방식임

## AR(1) model

$$Z_t - \mu = \phi (Z_{t-1}- \mu) + \varepsilon_t,\quad \varepsilon_t \sim WN(0, \sigma^2)$$

* 가정 : $\varepsilon_t$ 와 $Z_k , k<t$는 독립이다.

* MMSE 예측값 : $Z_n(l) = \mu + \phi^l(Z_n-\mu)$ 너무 당연한 이야기...

*  단 정상시게열을 만족하려면 $\to |\phi| <1$

* 또한 $l$ 이 커지면 $Z_n(l) \approx \mu$ 이러한 이유때문에 시계열 모형은 가까운 미래를 예측하는데 사용한다.

* 예측오차

$\begin{eqnarray*}\varepsilon_n(l) &=& Z_{n+l}-Z_{n}(l)\\ \\
                                   &=& Z_{n+l} - [ \mu + \phi^l (Z_n -\mu)] \\ \\ 
                                   &=& \varepsilon_{n+l}+ \phi \varepsilon_{n+l-1} + \dots \phi^{l-1}\varepsilon_{n+1} \\ \\
                                   &=& \sum_{i=0}^{l-1} \phi^i \varepsilon_{n+l-i}\end{eqnarray*} $

$\begin {eqnarray*}\therefore \quad E(\epsilon_n(l))=0\quad Var(\epsilon_n(l))= \frac {\sigma^2(1-\phi^{2l})}{1-\phi^2} \leq \frac {\sigma^2}{1-\phi^2}\end{eqnarray*}$

* 즉 $l$ 이 커지게되면 $Var(\epsilon_n(l)) \approx \frac {\sigma^2}{1-\phi^2}$이며 따라서 멀리있는 시점을 예측할 수록 불확실성이 커짐을 알 수있다.

* 에측구간

$$ \hat {Z}_n \pm z_{\alpha/2}\hat {\sigma} \sqrt {\frac {1-\hat{\phi}^{2l}}{1-\hat {\phi}^2}}$$

## MA(1) model

* MMSE 예측값

$$Z_t-\mu = \varepsilon_t-\theta \varepsilon_{t-1}$$

$$ Z_n(l) = \begin{cases} \mu - \theta \varepsilon_n,\quad l=1 \\\\ 
                          \mu, \quad l>1\end{cases}$$

* MA(1) 같은 경우 신뢰구간이 2번째 시점부터 예측오차가 커진다. 그러나 그 이후 계속 일정히 지속됨

## 2. IMA(1,1)

```r
z <- scan("eg9_5.txt")

dt <- tibble( t = 1:length(z),
                  z = z)
```

```r
p3 <- ggplot(dt, aes(t, z)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+ ggtitle('')+
  theme_bw()

p1 <- ggAcf(dt$z) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$z) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

```r
adfTest(dt$z, lags = 0, type = "c")
adfTest(dt$z, lags = 1, type = "c")
adfTest(dt$z, lags = 2, type = "c")
```

* 단위근 검정결과 단위근을 가지고 있다고 판단됨 즉, 차분이 필요하다고 판단.

```r
dt$diff_z  <-  c(0,diff(z))
```

```r
p3 <- ggplot(dt, aes(t, diff_z)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+
  theme_bw()

p1 <- ggAcf(dt$diff_z) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$diff_z) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

```r
adfTest(dt$diff_z, lags = 0, type = "nc")
adfTest(dt$diff_z, lags = 1, type = "nc")
adfTest(dt$diff_z, lags = 2, type = "nc")
```

* 귀무가설을 기각하여 더 이상의 단위근은 없다고 판단됨.

* 아래 2개의 모형은 근사한 모형임을 보여주는 것임, 즉 차분을 앞서 하든 모형적합에서 하든 비슷한 결과이다.

```r
fit1 <- arima(dt$diff_z, order=c(0,0,1), include.mean = F)
fit1
##### Wt=(1-B)Zt, Wt=et-theta*e(t-1)  
#####      => (1-B)Zt=et-theta*e(t-1) ## ARIMA(0,1,1)
```

```r
fit <- arima(dt$z, order=c(0,1,1), include.mean = F)
fit
# (1-B)Zt=et-theta*e(t-1) ## ARIMA(0,1,1)
```

* 잔차분석

```r
dt$resid <- as.numeric(resid(fit))
```

```r
p3 <- ggplot(dt, aes(t, resid)) + 
  geom_line(col='steelblue') +
  geom_hline(yintercept = 0, col='grey', lty=2)+
  xlab("")+ylab('')+ ggtitle('Time serise plot of Residual')+
  theme_bw()

p1 <- ggAcf(dt$resid) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$resid) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```
