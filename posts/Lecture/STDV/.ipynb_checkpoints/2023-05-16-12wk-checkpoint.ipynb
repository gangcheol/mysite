## 순서 1 : 지역구 그리기

````python
local_distriction_jsonurl='https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2018/json/skorea-municipalities-2018-geo.json'
global_distriction_jsonurl='https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2018/json/skorea-provinces-2018-geo.json'
````

`-` json 파일 load $\to$ 복잡한 형채의 `dic` 형태로 저장되어 있음

````python
local_dict = json.loads(requests.get(local_distriction_jsonurl).text) ## 시군구별 json 파일
global_dict = json.loads(requests.get(global_distriction_jsonurl).text) ## 도별 json 파일
````

### 연습 : 덕진구와 완산구를 찾아보자.

`-` 지역 이름을 가지는 리스트를 생성

````python
_lst = [_features[i]["properties"]["name"] for i in range(250)]
````

````python
_lst[:10]
````

`-` index 함수 이용

````python
_lst.index("전주시덕진구")
````

````python
_lst.index("전주시완산구")
````

````python
_keywords = ["전주시덕진구","전주시완산구"]
_jj = [_lst.index(keyword) for keyword in _keywords]
_jj
````

`-` `view` vs `copy`

* 딕셔너리의 경우 copy를 명시하지 않으면 두 개의 객체가 동시의 변경된다.

* 따라서 딕셔너리를 이용할 경우 카피를 명시하고 사용하자

#### 시각화

````python
df = pd.DataFrame({'key':['전주시덕진구','전주시완산구'],'value':[20,30]})
df
````

````python
local_dict2 = local_dict.copy()
````

````python
local_dict2["features"] = [local_dict["features"][j] for j in _jj]
````

````python
m = folium.Map([35.84195368311022, 127.1155556693179],zoom_start=11,scrollWheelZoom=False)
choro = folium.Choropleth(
    data=df, 
    geo_data=local_dict2,
    columns=['key','value'],
    key_on = 'feature.properties.name'
)
choro.add_to(m) 
m
````

## 요약

`-` `folium.Choropleth`의 사용법

* 2개의 데이터를 연결해야함 : 하나는 일반 데이터 프레임, 하나는 json에서 나온 dict
* 데이터 연결을 위해서는 공유가능한 연결의 매체가 필요하다.
* 연결의 매개체는 df-key, local_dct2["],`local_dict2['features'][?]['properties']['name']`을 이용했다.
* 연결의 매개체를 `name` 이아닌 다른 것들도 이용할 수 있다.

````python
m = folium.Map([35.84195368311022, 127.1155556693179],zoom_start=11,scrollWheelZoom=False)

choro = folium.Choropleth(
    data=df,  ## data1 
    geo_data=local_dict2, ## data2 (이 시점에서 폴리곤에 대한 정보가 choro 인스턴스에 전달) 
    columns=['key','value'], ## data1에서 중요한것들을 나열. 항상 [key,value]의 조합으로 써야한다. 
        # 이때 ['key','value']는 [data2와의 매개체,지도에서 색의 단계를 표현하는 변수]로 해석가능
    key_on = 'feature.properties.name' ## data2에서 중요한것: 즉 data1과의 연결매개체 
)
choro.add_to(m)
m
````

## Choropleth Map of Plotly

````python
import plotly.express as px
from IPython.display import HTML

df = px.data.election()
geojson = px.data.election_geojson()

fig = px.choropleth_mapbox(df, ## 데이터 1 
                           geojson=geojson, ## 데이터 2
                           color="Bergeron",
                           locations="district", featureidkey="properties.district",
                           center={"lat": 45.5517, "lon": -73.7073},
                           mapbox_style="carto-positron", zoom=9)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
HTML(fig.to_html(include_plotlyjs='cdn',include_mathjax=False, config=dict({'scrollZoom':False})))
fig.show()
````

# 판다스 백엔드

````python
import numpy as np 
import pandas as pd 
import warnings
import matplotlib.pyplot as plt
%matplotlib inline
````

`-` 핸드폰 점유율

````python
df = pd.read_csv('https://raw.githubusercontent.com/kalilurrahman/datasets/main/mobilephonemktshare2020.csv')
df
````

## matplotlib : 2개의 y를 겹처그리기

````python
df.plot.bar(x="Date", y= ["Samsung","Apple"],figsize=(10,5))
````

## 막대 width 조정

````python
df.plot.bar(x="Date",y = ["Samsung","Apple"],figsize=(10,5),width = 0.8)
````

## barh = 그래프 가로로

````python
df.plot.barh(x='Date',y=['Samsung','Apple'],figsize=(5,10))
````

## subplot 적용

````python
df.plot.bar(x='Date',figsize=(15,10), subplots=True, layout=(4,4), legend=False)
````

## 평균시각화

````python
df.melt(id_vars = "Date")
````

````python
df.melt(id_vars = "Date").groupby("variable")\
    .agg(np.mean).reset_index().\
    plot.bar(x="variable",y="value",legend=False)
````

## sorting 후 시각화

````python
df.melt(id_vars = "Date").groupby("variable")\
    .agg(np.mean).reset_index().\
    sort_values("value",ascending=False).\
    plot.bar(x="variable",y="value",legend=False)
````

## plotly

````python
df.melt(id_vars="Date").groupby("variable").\
        agg(np.mean).reset_index().\
        sort_values("value",ascending=False).\
        plot.bar(backend="plotly",x="variable",y="value")
````

### 회사별 색깔 입히기

````python
df.melt(id_vars='Date').\
plot.bar(x='Date',y='value',color='variable',backend='plotly')
````

### 삼성, 애플, Huawel만 표시

````python
df.melt(id_vars="Date").\
            query("variable == 'Samsung' | variable == 'Apple' | variable == 'Huawei'").\
             plot.bar(x="Date",y="value",color="variable",backend="plotly",barmode="group")
````

### + text 수치

````python
df.melt(id_vars='Date').query("variable=='Samsung' or variable=='Apple' or variable=='Huawei'" ).\
plot.bar(x='Date',y='value',color='variable',
            backend='plotly',barmode='group',text='value')
````

### + fact_col

````python
df.melt(id_vars='Date').query("variable=='Samsung' or variable=='Apple' or variable=='Huawei'" ).\
plot.bar(x='Date',y='value',color='variable',backend='plotly',facet_col='variable')
````

### + facet_row 

````python
df.melt(id_vars='Date').query("variable=='Samsung' or variable=='Apple' or variable=='Huawei'" ).\
plot.bar(y='Date',x='value',color='variable',backend='plotly',facet_row='variable',height=700)
````
