# 회귀분석

`1`. 다음과 같은 자료가 주어졌다고 하자

```python
x =  np.array([0.00983, 0.01098, 0.02951, 0.0384 , 0.03973, 0.04178, 0.0533 ,
               0.058  , 0.09454, 0.1103 , 0.1328 , 0.1412 , 0.1497 , 0.1664 ,
               0.1906 , 0.1923 , 0.198  , 0.2141 , 0.2393 , 0.2433 , 0.3157 ,
               0.3228 , 0.3418 , 0.3552 , 0.3918 , 0.3962 , 0.4    , 0.4482 ,
               0.496  , 0.507  , 0.53   , 0.5654 , 0.582  , 0.5854 , 0.5854 ,
               0.6606 , 0.7007 , 0.723  , 0.7305 , 0.7383 , 0.7656 , 0.7725 ,
               0.831  , 0.8896 , 0.9053 , 0.914  , 0.949  , 0.952  , 0.9727 ,
               0.982  ])
```

아래의 수식에 따라 $y_i$를 생성

* $y_i = 2+3x_i+\varepsilon_i, \quad \varepsilon_i \overset{iid}\sim N(0.1)$

```python
e = np.random.normal(size=len(x))
y = 2 + 3*x +e
```

```python
plt.plot(x,y,"o")
```

`2` 아래의 표에 맞는 형태로 추세선을 시각화해라

$$\hat y = ax + b$$

|a|b|
|:---:|:---:|
|1|0|
|2.5|2|
|3|2|

```python
y1 = x
y2 = 2.5*x+2
y3 = 3*x+2
```

```python
plt.plot(x,y1,label=r"y=x")
plt.plot(x,y2,label=r"y=2.5x+2")
plt.plot(x,y3,label=r"y=3x+3")
plt.plot(x,y,"--o",label ="y",alpha=0.3)
plt.legend()
```

`3` 각 추세선의 MSE를 계산

```python
np.mean((y-y1)**2),np.mean((y-y2)**2),np.mean((y-y3)**2)
```

`4` 다음의 수식을 이용하여 $\beta$를 계산

$$\hat {\beta} = (X^{T}X)^{-1}X^{T}y$$

```python
X = np.stack([np.ones(len(x)),x],axis=1)
```

```python
beta  = np.linalg.inv(X.T@X)@X.T@y
```

```python
beta
```

`5` 추정된 베타를 이용하여 추세선을 시각회

```python
hat_y = X@beta
```

```python
plt.plot(x,y,"o",alpha=0.3,label=r"$y$")
plt.plot(x,hat_y,"--",label=r"$\hat y$")
plt.legend()
```

## (4) 삼성전자 주가

```python
start_date = "2023-01-01"
end_date = "2023-05-02"
y = yf.download("005930.KS", start=start_date, end=end_date)['Adj Close'].to_numpy()
```

```python
plt.plot(y)
```

아래와 같은 변환을 이용하여 $y$를 $\hat y$ 변환하고 결과를 시각화해라

* $\hat y_1 = \frac 14 (3y_1+y_2)$

* $\hat y_i = \frac 14 (y_{i-1}+2y_i+y_{i+1})$

*  $\hat y_n = \frac 14 (y_{n-1}+3y_n)$

### sol

```python
T = len(y)
M = (np.eye(T) + np.array([abs(i-j)<2 for i in range(T) for j in range(T)]).reshape(T,T))/4
M[0,0] = 3/4; M[-1,-1]= 3/4 
plt.plot(y,label=r'$y$')
plt.plot(M@y,'--',label=r'$M@y$')
plt.legend()
```

## (5) (4)의 변환을 50회 반복적용하고 결과를 시각화

```python
plt.plot(y,label=r'$y$')
plt.plot(np.linalg.matrix_power(M,50)@y,'--',label=r'$M^{50}@y$')
plt.legend()
```
