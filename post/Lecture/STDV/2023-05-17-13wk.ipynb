# Pie chart

`-` 각 회사별 비율을 표시하여 그리기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kalilurrahman/datasets/main/mobilephonemktshare2020.csv')
df.head()
```

`-` 아래와 같은 형태는 우리가 원하는 형태의 시각화가 아님.

## 문제 1 : 날짜인덱스 x

```python
df.plot.pie(y="Samsung")
```

## 문제 2: 날짜 인덱스 O, 목적 x

```python
df.set_index("Date").plot.pie(y="Samsung")
```

### 문제 3 : 드디어 정상적인 플랏, but 특정 기간만 표시됨

```python
df.set_index("Date").T.plot.pie(y="2019-10",legend=False)
```

### 완성

```python
df.set_index("Date").T\
    .plot.pie(legend=False,subplots=True,layout=(4,4),figsize=(20,20))
```

# Boxplot

## 예제 1 (matplotlib)

`-` 월별 점유율의 차이를 구한 후 boxplot을 그려보자

```python
df.set_index("Date").diff().dropna()
```

```python
df.set_index('Date').diff().dropna()\
    .boxplot(figsize=(20,5))
```

`-` 삼성의 경우는 점유율 변동이 거의 없다.

`-` 파이플랏으로 변동폭을 잡기 어려울 때는 아래처럼 변동의 차이를 boxplot로 그려 확인할 수 있다.

`-` 음수값에 값이 몰려있으먄 점유율이 전체적으로 떨어지고 있다는 판단을 내릴 수 있다.

`-` 또한, 중앙값을 보고 전체적인 추세를 짐작할 수 있다.

## 예제 2 (plotly)

`-` 위와 동일한 그림을 plotly backend로 그릴 수 있음

```python
import plotly.express as px
```

```python
df.set_index("Date").diff().\
  dropna().boxplot(backend = "plotly")
```

### 회사별 색깔 추가

```python
df.set_index("Date").diff().dropna().\
        melt().boxplot(backend="plotly",color="variable")
```

## 예제 3 (plotly)

```python
df = px.data.tips()
df.head()
```

### 팁은 대체로 얼마나 받을까? 

```python
df.plot.box(backend = "plotly", y = "tip")
```

### 시간대, 성별 구분

```python
df.plot.box(backend = "plotly",
            x = "time",
            y = "tip", color = "sex")
```

`-` 점심은 남자가 더 많이 받는 것 같구, 저녁은 비슷하다. (중앙값을 보고 판단)

`-` boxplot의 다점은 히스토그램처럼 해당 구간에 점들의 밀집도를 보기 힘들다는 것이다. 즉, 전체적인 경향만 알 수 있다.

```python
df.plot.box(backend='plotly',
            y='tip',x='time',
            color='sex',points='all')
```

`-` point를 그려본 결과 여자보다 남자가 팁을 받는 횟수 자체가 많다. (얼마를 주는지는 알 수 없지만....)


`-` boxplot의 아웃라이어가 사라지고, 데이터를 더 리치하게 볼 수 있다!

### 시간대 + 성별 + 요일

```python
df.plot.box(backend = "plotly",
            x = "time",
            y = "tip",
            color = "sex",
            facet_col = "day",
            points = "all")
```

### 시간대 + 성별 + 요일 + 흡연유무

```python
df.plot.box(backend = "plotly",
            x = "time",
            y = "tip",
            color = "sex",
            facet_col = "day",
            facet_row = "smoker",
            points = "all")
```

```python
 df = pd.DataFrame(
    [(np.random.normal(size=(1000,i),loc=0,scale=1)**2)\
      .sum(axis=1) for i in range(3,11)]
 ).T
```

 ## hist

```python
df.plot.hist(alpha = 0.2, bins = 50, 
             subplots = True, 
             layout=(2,4),
             figsize = (15,5))
```

### kde plot

`-` bw_method : 조금 지글지글한? 아래 그래픛처럼 스무딩 효과를 얼만큼 줄 것인지 결정한다.

# area plot

`-` 주로 다루는 plot은 아님

## matplotlib

```python
df = pd.read_csv('https://raw.githubusercontent.com/kalilurrahman/datasets/main/mobilephonemktshare2020.csv')
df.head()
```

```python
df.set_index("Date").plot.area(alpha=0.6,figsize=(15,5))
```

`-` 시간에 따른 점유율의 변화를 보기에 적절한 것 같다? $\to$ 내 생각엔 그렇지 않음..

`-` 각 회사의 y축 폭은 점유율을 뜻함.

## plotly

```python
df.melt(id_vars="Date").\
  plot.area(backend = "plotly",
            x = "Date",
            y = "value",
            color = "variable")
```

`-` 끔직할 정도의 데이터이다.

`-` 전국 단위 제거

```python
df.iloc[1:,]
```

`-` 컬럼확인 결괴 멀티 인덱스임을 확인

```python
df.iloc[1:].set_index(("행정구역별","행정구역별")).head()
```

```python
df.iloc[1:].set_index(("행정구역별","행정구역별")).stack()
```

```python
df.iloc[1:].set_index(("행정구역별","행정구역별")).\
            stack().reset_index().head()
```

`-` 비율 시각화가 목적이므로 비율만 선택하자.

```python
mat_data = df.iloc[1:].set_index(("행정구역별","행정구역별")).\
            stack().reset_index().\
            query("level_1 == '비율'").\
            set_index(("행정구역별","행정구역별")).T.iloc[1:,:]
```

```python
mat_data
```

### applymap

`-` 전처리

```python
mat_data = mat_data.applymap(lambda x : float (x) if x !="-" else None)
```
