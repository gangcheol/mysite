# 시벤코정리

`-`하나의 은닉층을 가지고 있고 충분한 수의 노드가 보장되면, 대부분의 함수를 적당히 근사시킬 수 있다.

## 지나시간 논리전개

`-` 아이디어 :  아이디어: linear -> relu -> linear (-> sigmoid) 조합으로 꺽은선으로 표현되는 underlying 을 표현할 수 있었다.

* 실제로 꺽은선이 아니라도 꺽은선의 조합인 곡선도 표현할 수 있다.

* X가 (n, 2)인 경우도 가능했다. (이 경우 꺽인선이 꺽인 평면이 된다.??
)

* 아이디어에 해당하는 용어정리: 이 구조가 x->y 로 바로 가는 것이 아니라 x->(u1->v1)->(u2->v2)=y 의 구조인데 이러한 네트워크를 하나의 은닉층을 포함하는 네트워크라고 표현한다. (이 용어는 이따가..)

## 시벤코 정리

**universal approximation thm** : (범용근사정리,보편근사정리,시벤코정리), 1989

> 하나의 은닉층을 가지는 "linear -> sigmoid -> linear" 꼴의 네트워크를 이용하여 세상에 존재하는 모든 연속함수를 원하는 정확도로 근사시킬 수 있다.(계수를 잘 추정한다면)

`-` 이해가 안되는 부분

*  그렇게 잘 맞추면 1989년에 세상의 모든 문제를 다 풀어야 한거 아닌가?

* 요즘은 "linear -> sigmoid -> linear"이 아닌 "linear -> relu -> linear" 조합으로 많이 쓰던데

* 요즘은 하나의 은닉층만 포함하는 네트워크는 잘 안쓰지 않나?

`-` 약간의 의구심이 있지만 해당 이론에 따르면 아래와 같다.

* $\bf {X} : (n,p)$ 꼴의 입력에서 $\bf{y} : (n,1)$ 꼴의 출력으로 향하는 맵핑을 "linear -> relu -> linear" 와 같은 네트워크를 이용해 **"근사"**시킬 수 있다.

```python
l1.weight.data = torch.tensor([[-5.00],[5.00],[-5.00],[5.00]])
l1.bias.data = torch.tensor([0.00, 20.00, 20.00, 0])
l2.weight.data = torch.tensor([[1.00,  1.00, 2.50,  2.50]])
l2.bias.data = torch.tensor([-1.0-2.5])
```

```python
plt.plot(l2(a1(l1(x))).data)
```

`-` 2개의 시그모이드를 우현히 잘 겹합하면 아래와 같은 함수 $h$ 를 만들 수 있다.

```python
h = lambda x : torch.sigmoid(200*(x+0.5)) + torch.sigmoid(-200*(x-0.5))-1.0
```

```python
plt.plot(x,h(x))
plt.title("$h(x)$")
```

`- `위와 같은 함수 $h$ 를 활성화함수로 하고 $m$ 개의 노드를 가지는 은닉층을 생각해보자. 이러한 은닉층을 사용한다면 전체 네트워크를 아래와 같이 표현할 수 있다.

$\underset{(n,1)} {\bf{X}} \overset{l_1}{\to} \underset{(n,m)}{u^{(1)}} \overset{h}{\to} \underset{(n,m)}{v^{(1)}} \overset{l_2}{\to} \underset{(n,1)}{\bf{\hat{y}}}$

그리고 위의 네트워크와 동일한 효과를 주는 아래의 네트워크가 항상 존재함

$\underset{(n,1)} {\bf{X}} \overset{l_1}{\to} \underset{(n,2m)}{u^{(1)}} \overset{sig}{\to} \underset{(n,2m)}{v^{(1)}} \overset{l_2}{\to} \underset{(n,1)}{\bf{\hat{y}}}$

`-` 첫 번쨰 루틴을 증명하면 두 번쨰 루틴을 증명하는 것과 같다.
  
  * h = sig + sig 이므로

`-` $h(x)$를 활성화함수로 가지는 네트워크를 설계

```python
class MyActivation(torch.nn.Module): ## 사용자정의 활성화함수를 선언하는 방법
    def __init__(self):
        super().__init__() 
    def forward(self, input): ## x에서 y로 가는 함수!
        return h(input) # activation 의 출력 
```

```python
a1=MyActivation()
# a1 = torch.nn.Sigmoid(), a1 = torch.nn.ReLU() 대신에 a1 = MyActivation()
```

```python
plt.plot(x,a1(x)) 
```

`-` **히든레이어가 1개의 노드를 가지는 경우** $\to$ 총 16개의 경우를 그린거임, 16번의 새로운 계수를 이용한 거!

```python
torch.manual_seed(202150256)
fig, ax = plt.subplots(4,4,figsize=(12,12))
for i in range(4):
    for j in range(4):
        net = torch.nn.Sequential(
            torch.nn.Linear(1,1),
            MyActivation(),
            torch.nn.Linear(1,1)
        )
        ax[i,j].plot(x,net(x).data,'--')
```

## 예제 1 (sin, exp)

`-` 아래와 같은 underlying을 가지는 $\hat {y}$를 만들어보자

```python
torch.manual_seed(202150256)
x = torch.linspace(-10,10,200).reshape(-1,1)
underlying = torch.sin(2*x) + torch.sin(0.5*x) + torch.exp(-0.2*x)
eps = torch.randn(200).reshape(-1,1)*0.1
y = underlying + eps 
plt.plot(x,y,'o',alpha=0.5)
plt.plot(x,underlying,lw=3)
plt.legend(["y","underlying"])
```

```python
h = lambda x: torch.sigmoid(200*(x+0.5))+torch.sigmoid(-200*(x-0.5))-1.0
class MyActivation(torch.nn.Module): ## 사용자정의 활성화함수를 선언하는 방법
    def __init__(self):
        super().__init__() 
    def forward(self, input):
        return h(input) 
```

```python
net= torch.nn.Sequential(
    torch.nn.Linear(1,2048),
    MyActivation(),
    torch.nn.Linear(2048,1)
)
loss_fn = torch.nn.MSELoss()
optimizr = torch.optim.Adam(net.parameters()) 
```

```python
for epoc in range(200):
    ## 1 
    yhat = net(x) 
    ## 2 
    loss = loss_fn(yhat,y)
    ## 3 
    loss.backward()
    ## 4 
    optimizr.step()
    optimizr.zero_grad()
```

```python
plt.plot(x,y,'o',alpha=0.2)
plt.plot(x,underlying,lw=3)
plt.plot(x,net(x).data,'--')
plt.legend(["y","underlying","net(x)"])
```

***Layer을 세는 방법***

* 정석 : 학습가능한 파라메터가 몇층으로 있는지

***Hidden Layer*** 의 수를 세는 방법

*  Layer의 수 = Hidden Layer의 수 + 출력층의 수 = Hidden Layer의 수 + 1

* node의 개념 : $u \to v$ 로 가는 쌍을 간단히 노드라는 개념을 이용하여 나타낼 수 있음

## 굳이 GPU에 넘겨야 하는가?

`-` 데이터 셋을 짝홀로 나누어서 번갈아가며  GPU에 올렸다 내렸다하면 안되나?

`-` 알고리즘

1. 데이터를 반으로 나눈다.

2. 짝수 관측치와  net의 파라미터를 GPU에 올린다.

3. 학습 수행

4. 짝수 관측치를 GPU 메모리에서 내린다. 그리고 홀수 관측치를 GPU메모리에 올린다.

5. 학습 수행

6. 홀수 관측치를 GPU메모리에서 내린다. 그리고 짝수 관측치를 GPU메모리에 올린다.

7. 반복....

`-` 미니배치에서 안쓰는 학습

```python
torch.manual_seed(202150256)
net = torch.nn.Sequential(
    torch.nn.Linear(784,32),
    torch.nn.ReLU(),
    torch.nn.Linear(32,1),
    torch.nn.Sigmoid()
)
loss_fn = torch.nn.BCELoss()
optimizr = torch.optim.Adam(net.parameters())
```

```python
for epoc in range(70): 
    ## 1 
    yhat = net(X) 
    ## 2 
    loss= loss_fn(yhat,y) 
    ## 3 
    loss.backward()
    ## 4 
    optimizr.step()
    optimizr.zero_grad() 
```

```python
torch.sum((yhat>0.5) == y) / len(y) 
```

`-` 미니배치를 쓰는 학습 (GPU에 올리고 내리는 과정은 생략)

```python
torch.manual_seed(202150256)
net = torch.nn.Sequential(
    torch.nn.Linear(784,32),
    torch.nn.ReLU(),
    torch.nn.Linear(32,1),
    torch.nn.Sigmoid()
)
loss_fn = torch.nn.BCELoss()
optimizr = torch.optim.Adam(net.parameters())
```

```python
for epoc in range(10):
    for xx,yy in dl: ## 7번
        ## 1
        #yhat = net(xx)
        ## 2 
        loss = loss_fn(net(xx),yy) 
        ## 3 
        loss.backward() 
        ## 4 
        optimizr.step()
        optimizr.zero_grad()
```
