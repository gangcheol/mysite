# 예제 2.1 선형추세모형의 적합 예제

```r
library(tidyverse)
library(TTR)
library(lmtest)
library(astsa)
```

```r
setwd("C:\\Users\\lee\\Desktop\\고급시계열 분석\\제5판_시계열분석_프로그램\\제5판 시계열분석 data")
```

```r
population <- scan("population.txt")
```

```r
options(repr.plot.res=200,repr.plot.width=10)
plot.ts(population)
```

* 시도표를 그려본 결과 뚜렷한 증가 추세가 보인다.
* 또한 미세하게 곡선의 형태를 가지므로 1차 선형추세모형과 2차 추세모형을 각각 적합해보기로 한다.

```r
t <- 1:length(population)
t2 <- t^2
```

```r
fit1 <- lm(population~t)
fit2 <- lm(population~t+t2)
```

```r
summary(fit1); summary(fit2) 
```

* 두 모델의 회귀계수와 모형이 모두 유의하다는 결과가 나왔으며 결정계수값을 보았을 때 높은 설명력을 보여주고 있다.

```r
par(mfrow=c(1,2))
plot(fit1,1); plot(fit2,1)
```

* 그러나 잔차 plot을 그려본 결과 선형추세 모형의 경우 시간변수의 제곱항이 필요한 것을 알 수 있다.

* 2차 추세모형의 잔차 plot을 살펴보았을 때 오차항간의 자기상관성이 존재해보이며 또한 잔차의 변동폭이 넓어지는 것을 확인할 수 있다.

* 따라서 기존의 자료의 로그변환을 하여 이분산성을 없앤 후 다시 모형을 적합해보자.

```r
log_data <- log(population)
fit3 <- lm(log_data~t+t2)
summary(fit3)
```

```r
plot(fit3,1)
```

* 변동폭은 어느 정도 안정화 된 것 같으나 여전히 잔차들이 시간에 따른 상관관계를 갖고 있는 것처럼 보인다.

```r
?dwtest
```

```r
dwtest(fit3,alternative="two.side")
```

* 실제로 더비왓슨통계량의 근거한 p-value 값을 살펴본 결과 오차항간의 자기상관성이 존재함을 보였다.

* 이 경우 오차들의 공분산을 이용한 일반화 최소제곱법을 이용하여 모형을 적합시키거나 자기회귀오차모형을 적합시키는 방법이 있다.

# 예제 2.2 추세성분과 계절성분을 동시에 갖는 모형 적합 예제

```r
depart <- scan("depart.txt")
depart <- ts(depart, start =c(1984,1),frequency = 12)
plot(depart)
```

* 시도표를 그려본결과 추세성분과 계절성분이 관측되었으며, 시간의 흐름에 따라 변동폭이 커짐을 알 수 있다.

* 따라서 로그변환으로 변동폭을 균등하게 변환

```r
log_depart <- log(depart)
plot(log_depart)
```

* 변동폭이 안정화 되었으며 지시함수를 사용한 모형을 적합시킨다.
* 또한 모수들 사이의 제약조건으로 절편이 없는 모형을 적합시킴

```r
trend <- time(log_depart)-1984 ## 추세성분
month <- factor(cycle(log_depart)) ## 계절성분
reg <- lm(log_depart~0+trend+month) ## 절편이 없는 모형을 생성
summary(reg)
```

```r
dwtest(reg, alternative = "two.sided")
```

* 오차항간의 자기상관성이 발견되 적합한 모형으로 채택할 수 없음.

* 또한 ACF와 PACF를 시각화 한결과 특정 지점에서 자기상관성이 존재함을 알 수 있음

```r
acf2(resid(reg))
```
