## 1) 

적절한 추세 모형을 적합시킨 후 잔차분석을 하여라.

### Solution

```r
fit1 <- lm(z~t,data=data1)
summary(fit1)
```

산출된 회귀식은 $\hat z = 0.099619\times t + 0.818932$로 산출되었으며 $F$ 통계량의 근거한 $p-value$값을 보았을 때 적합된 모형은 통계적으로 유의하다.

결정계수값을 살펴본 결과 해당 모형은 전체 변동의 약 80%를 설명하고 있다.

```r
options(repr.plot.res=200,repr.plot.width=10,repr.plot.height=7)
par(mfrow=c(1,2))
for (i in 1:2)
{
  plot(fit1,i)
}
dwtest(fit1,alternative="two.sided")
```

잔차분석결과 몇몇 이상치로 추측되는 관측치들이 발견되었으나. 특정한 패턴이나 이분산성은 관측되지 않는다.

또한 `qqplot`을 통해 정규성을 확인한 결과 잔차의 분포가 정규분포에 가까운 형태임을 알 수 있었으며 자기상관성 검토 결과 오차항간의 자기 상관성 또한 보이지 않았다.

## 2) 

`1)` 에서 적합한 추세모형에서 1~10 시차 후의 예측값을 구하여라.

### Solution

$\hat Z(l) = \hat {\beta_0}+\hat {\beta_1}\times(t+l)$ 이므로

```r
beta <- fit1$coefficients
pred <- beta[1]+beta[2]*(101:110)
```

```r
options(repr.plot.res=200,repr.plot.width=8,repr.plot.height=7)
plot(1:100,fit1$fitted.values,
     xlim=c(1,110),ylim=c(1,15),type="l",lwd=2,
     main ="1~10시차 후 예측값 갱신")
points(x=101:110,pred,type="l",col="red",lwd=3)
legend("topleft",
       pch="l",c("z예측값","1~10시차 후 예측값"),col=c("black","red"))
pred
```

## 3) 

적절한 평활법을 적용한 후 잔차분석을 하여라.

### Solution

* 시도표를 그려본결과 추세성분과 계절성분, 불규칙성분이 관측되었다.
* 따라서 **Winters의 계절지수평활법**을 사용하여 평활을 진행하겠다.
* 변동의 폭은 시간의 흐름에 관계없이 일정하므로 가법계절모형을 적용

```r
options(repr.plot.res=200,repr.plot.width=8,repr.plot.height=5)
data1_ts <- ts(data1$z,frequency = 12)
plot.ts(data1_ts,type="l",lwd=2,xlim=c(1,10))
abline(v=seq(0,9,1),lty="dashed",col="red",lwd=2)
```

```r
fit2 <- HoltWinters(data1_ts,seasonal = "additive")
fit2
```

```r
ts.plot(resid(fit2), ylab="residual", 
        main="가법모형의 예측오차",lwd=2); abline(h=c(0,3,-3),lwd=2,col="red",lty="dashed")
```

* 잔차그림을 그려본 결과 앞부분과 뒷부분에서 변동이 큰 값들이 관측되었으나 잔차의 분산이 이분산성이라고 판단되지는 않는다.

# 2

`usapass.txt` 는 미국 월별 비행기 승객 수(단위 : 천 명)의  시계열자료이다. log 변환 후 아래의 분석을 수행하시오.

## 1)

왜 log 변환이 필요한지에 대해  간단히 설명하여라.

### Solution

```r
data2 <- scan("usapass.txt")
```

* 시도표를 보았을 때 추세성분, 계절성분, 불규칙성분이 모두 관측되었다.
* 또한 시간의 흐름에 따라 변동폭이 커지는 이분산성을 보이기 때문에 로그변환과 같은 방법을 이용하여 접근해야한다.
* 실제로 변환시 기존보다 안정된 변동폭을 보여주었다.

```r
par(mfrow=c(1,2))
log_d <- log(data2)
plot.ts(data2,main="변환 전");plot(log_d,type="l",main="변환 후")
```

## 2)

적절한 추세모형 적합 후 잔차분석을 하여라.

### Solution

```r
t <- 1:length(log_d)
fit3 <- lm(log_d~t)
par(mfrow=c(1,2))
plot(fit3,1);plot(fit3,2)
dwtest(fit3,alternative="two.sided")
```

* 잔차분석결과 빨간색 잔차 추세선이 곡선의 형태로 띄어 2차항 등의 추가적인 요소가 필요하다고 판단된다.

* 또한 몇몇 이상치들이 관측되었다.

* 그러나 `qqplot`을 통해 자세히 정규성을 확인한 결과 잔차의 분포가 정규분포에 가까운 형태임을 알 수 있다.

* 오차항간의 자기상관성 검토 결과 오차항간의 양의 자기상관성이 존재함을 보였다.

## 3)

적절한 평활법을 적용한 후 잔차분석을 하여라.

### Solution

* 로그변환을 통해 생성된 데이터가 어느 정도 등분산임을 보이고, 계절성분과 추세성분이 관측되었기 때문에 앞선 문제와 동일하게 **Winters의 계절지수평활법**을 사용하겠다.

```r
data2_ts <- ts(log_d,frequency = 12)
fit4 <- HoltWinters(data2_ts,seasonal = "additive")
```

```r
ts.plot(resid(fit4), ylab="residual", 
        main="가법모형의 예측오차",lwd=2); abline(h=c(0,0.1,-0.1),lwd=2,col="red",lty="dashed")
```

* 잔차의 변동폭이 시간의 흐름에 따라 어느 정도 일정함을 보인다.

## 4)

* 적절한 분해법에 의해 각 성분을 분해하여 시계열 그림을 그려라.

### Solution

#### 1) 이동평균법 사용 : 가법모형

```r
decomp_data <- decompose(data2_ts)
plot(decomp_data)
```

* 이동평균법을 사용한결과 불규칙성분에서 미묘하게 체계적인 패턴이 보이는 곳들이 발견되었다. 
* 따라서 추세모형에 의한 분해법을 시행하여 한번 더 분해를 해보자.

#### 2) 추세모형에 의한 분해

```r
fit7 <- lm(log_d ~ t )  
summary(fit7)
```

* 추정된 추세성분 : $\hat {T_t} = 4.8131 + 0.01008\times t$

```r
trend <- fitted(fit7)
ts.plot(data2_ts, trend, col=1:2, lty=1:2, ylab="승객 수(단위 : 천 명)", xlab="time",
        main="log변환한 시계열과 분해법에 의한 추세성분",lwd=2)
legend("topleft", lty=1:2, col=1:2, c("ln(data)", "추세성분"),lwd=2)
```

```r
adjtrend <- data2_ts-trend
y <- factor(cycle(adjtrend))

fit8 <- lm(adjtrend~0+y)
summary(fit8)
```

* 추정된 계절 성분 : $S_t$

```r
fit8$coefficients
```

```r
seasonal <- fitted(fit8)
ts.plot(seasonal)
```

* 추정값 : $\hat {Z_t} = \hat{T_t}+\hat {S_t}$

```r
pred <- trend + seasonal

ts.plot(data2_ts, pred, col=1:2, lty=1:2, ylab="승객 수(단위 : 천 명)", xlab="time",
        main="log변환한 시계열과 분해법에 의한 추정값",lwd=2)
legend("topleft", lty=1:2, col=1:2, c("ln(data)", "추정값"),lwd=2)
```

```r
irregular <- data2_ts - pred
ts.plot(irregular,main="추세성분에 의한 분해 결과 : 불규칙 성분"); abline(h=0, lty=2,col="red",lwd=2)
dwtest(lm(irregular~ 1))
```

4)에서 추정된 불규칙성분을 통해 적용된 분해법이 적절했는지 논하여라.

### Solution

* 이동평균모형에 의한 분해법과 추세모형에 의한 분해법을 적용하였을 때 잔차 plot을 그려본 결과 미미한 패턴들이 발견되었다.
* 추세모형에 의한 분해의 경우 오차항간의 자기상관성 검토 결과 양의 자기상관성을 보였다,

# 3

확률과정 $Z_t = 1+0.9Z_{t-1}+\varepsilon_t,t=1,2\dots,100$으로 부터의 시계열 자료를 생성한 후 다음을 수행하라.

## 1)

${Z_t}$의 시계열 그림을 그려라.

### Solution

```r
z <- vector()
z[1] <- 1+0.9*10+rnorm(1)
for (i in 2:100) z[i]=1+0.9*z[i-1]+rnorm(1)
plot(z,type="l", main =expression(Z[t]==1+0.9*Z[t-1]+epsilon[t]))
```

## 2)

### Solution

SACF $\hat {\rho}_k,k=0,1,2\dots 10$을 구하여 표본상관도표를 그려라.

```r
z_acf <- acf(z,maxlag=10,plot=F)
plot(z_acf[0:10])
```

* 시차 6까지는 $Z_t$들은 자기상관관계가 존재한다고 볼 수 있다.

## 3)

### Solution

SPACF $\hat {\phi}_{kk},k =0,1,2\dots10$을 구하여 표본상관도표를 그려라.

```r
z_pacf <- pacf(z,plot=F)
plot(z_pacf[0:10])
```

* PACF 그래프를 시차 1지점에서 유의미한 상관성이 보인다.
* 차수가 1인 AR(1) 모형을 고려해볼 수 있다.

## 4,5)

($Z_t, Z_{t−1}$ ),  ($Z_t, Z_{t−2}$)의 산점도를 그리고, 이 산점도와 $\hat\rho_1,\hat\rho_2$의 관계를 설명하여라.

### Solution

```r
par(mfrow=c(1,2))
z_t_1 <- c(10,z[1:99])
z_t_2 <- c(10,z[1:98])
plot(x=z_t_1,y=z,main=expression(list(Z[t-1],Z[t])))
plot(x=z_t_2,y=z[2:100],main=expression(list(Z[t-2],Z[t-1])))
```

* 위 산점도는 각 시차간 값들의 상관관계 정도가 아래와 같다는 의미이다.

```r
z_acf[1:2]
```

* 실제로 산점도를 그려보았을 때도 왼쪽 그래프의 경우가 더 선형적인 모양을 띄었다.

# 4

$\varepsilon_t\sim_{iid}N(0,1)$ 을 따를 때, 다음의 확률 과정을 따르는 표본의 크기가 100인 시계열을 생성한 후, 각 모의실험 자료에 대하여 다음에 답하여라.  단 AR 혹은 ARMA 과정에서 AR 부분의 차수가 p인 경우 필요한 초기값은 해당하는 확률과정의 평균 𝜇로 놓기로 하자.


 모델 1: $Z_t = \varepsilon_t +0.7\varepsilon_{t-1}\sim MA(1)\quad\theta = -0.7$

모델 2: $Z_t - 0.8Z_{t-1}=\varepsilon_{t}\sim  AR(1) \quad \phi=0.8$

모델 3 : $Z_t -0.5Z_{t-1}= \varepsilon_t+0.3\varepsilon_{t-1} \sim ARMA(1,1) \quad \theta= -0.3,\,\phi=0.5,\,\delta=0$

* 앞 선 과제에서 모델 2와 모델 3은 각각 AR(1),ARMA(1,1) 모형임을 보였다.

* 모형 1의 경우 ($\theta=-0.7$)

$\Phi(B)\dot{Z}_t = \Theta(B)\varepsilon_t, \quad \Phi(B)=1,\Theta(B)=1-\theta B=1+0.7B $

($\theta=-0.7$)

MA(1)모형을 따른다.

## 1)

{$Z_t$}의 시계열 그림을 그려라.

### Solution

```r
z1 <- arima.sim(n=100,list(order=c(0,0,1),ma=0.7),rand.gen=rnorm)
z2 <- arima.sim(n=100,model=list(order=c(1,0,0),ar=0.8),rand.gen = rnorm)
z3 <- arima.sim(n=100,model=list(order=c(1,0,1),ar=0.5,ma=0.3),rand.gen = rnorm)

par(mfrow=c(1,3))
ts.plot(z1,main=(expression(MA(1)~~~theta==-0.7)))
ts.plot(z2,main=(expression(AR(1)~~~phi==0.8)))
ts.plot(z3,main=(expression(ARMA(1,1)~~~ phi==0.5~~~ theta==-0.3)))
```

## (2)

SACF $\hat {\rho}_k$와 SPACF $\hat {\phi}_{kk}\quad (k=0,1,2\dots10)$의 상관도표를 그려라.

### Solution

```r
options(repr.plot.res=250,repr.plot.width=8,repr.plot.hegiht=8)
par(mfrow=c(2,3))
acf(z1,lag.max=10);acf(z2,lag.max=10);acf(z3,lag.max=10)
pacf(z1,lag.max=10);pacf(z2,lag.max=10);pacf(z3,lag.max=10)
```

## (3)

2)의 상관도표가 각 확률과정에 대한 ACF, PACF의 이론적인 형태와 비슷한지 논하라.

### Solution

* 빨간점이 이론적인 값, 검은선이 실제로 계산된 값

#### 모델 1 : MA(1)

```r
k=1:10
par(mfrow=c(1,2))
ACF1 <- c(1,0.7/(1+0.7^2),rep(0,9))
PACF1 <-{(-(-0.7)^k)*(1-0.7^2)}/{1-(-0.7^(2*(k+1)))}

z1_acf <- acf(z1,plot=F)
plot(z1_acf[0:10])
points(x=0:10,y=ACF1,col="red",lwd=2)
z1_pacf <- pacf(z1,plot=F)
plot(z1_pacf[0:10])
points(x=1:10,y=PACF1,col="red",lwd=2)
```

* ACF의 경우는 이론적으로 시차 1 이후 부터는 0이 나와야하나 그렇지 않다.
* PACF의 경우도 이론적인 값과는 약간의 차이가 있음을 보인다.

#### 모델 2 : AR(1)

```r
par(mfrow=c(1,2))
k=1:10
ACF2 <- c(1,0.8^k)
PACF2 <- c(0.8,rep(0,9))
z2_acf <- acf(z2,plot=F)
plot(z2_acf[0:10])
points(x=0:10,y=ACF2,col="red",lwd=2)
z2_pacf <- pacf(z2,plot=F)
plot(z2_pacf[0:10])
points(x=1:10,y=PACF2,col="red",lwd=2)
```

* ACF와 PACF 둘다 이론적인 값에서 많이 벗어나 있다.

#### 모델 3: ARMA(1,1)

```r
par(mfrow=c(1,2))
h=2:10
ACF3 <- c(1,0.66187,(0.5^(h-1))*0.66187)
r_2 <- 0.5*0.66187
PACF3 <- c(0.66187,(r_2-0.66187^2)/(1-0.66187^2))
z3_acf <- acf(z3,plot=F)
plot(z3_acf[0:10])
points(x=0:10,y=ACF3,col="red",lwd=2)
z3_pacf <- pacf(z3,plot=F)
plot(z3_pacf[0:10])
points(x=1:2,y=PACF3,col="red",lwd=2)
```

* ACF, PACF 값이 이론적인 값에서 많이 벗어났다.
* ARMA(1,1) 모형은 PACF 계산값이 복잡해 시차 2까지만 값을 표시하였다.

## (4)

또한 표본의 크기를 증가시켜 모의실험을 한 후 3)의 결과 및 이론적인 형태와 비교하여라.

### Solution

```r
z1 <- arima.sim(n=10000,list(order=c(0,0,1),ma=0.7),rand.gen=rnorm)
z2 <- arima.sim(n=10000,model=list(order=c(1,0,0),ar=0.8),rand.gen = rnorm)
z3 <- arima.sim(n=10000,model=list(order=c(1,0,1),ar=0.5,ma=0.3),rand.gen = rnorm)
```

#### 모델 1 : MA(1)

```r
k=1:10
par(mfrow=c(1,2))
ACF1 <- c(1,0.7/(1+0.7^2),rep(0,9))
PACF1 <-{(-(-0.7)^k)*(1-0.7^2)}/{1-(-0.7^(2*(k+1)))}

z1_acf <- acf(z1,plot=F)
plot(z1_acf[0:10])
points(x=0:10,y=ACF1,col="red",lwd=2)
z1_pacf <- pacf(z1,plot=F)
plot(z1_pacf[0:10])
points(x=1:10,y=PACF1,col="red",lwd=2)
```

* 표본이 증가하니 이전보다 더 이론적인 값에 가까워진 경향을 볼 수 있었다.

#### 모델 2 : AR(1)

```r
par(mfrow=c(1,2))
k=1:10
ACF2 <- c(1,0.8^k)
PACF2 <- c(0.8,rep(0,9))
z2_acf <- acf(z2,plot=F)
plot(z2_acf[0:10])
points(x=0:10,y=ACF2,col="red",lwd=2)
z2_pacf <- pacf(z2,plot=F)
plot(z2_pacf[0:10])
points(x=1:10,y=PACF2,col="red",lwd=2)
```

* 마찬가지로 기존보다 더 이론적인 값에 가까운 값을 가지는 것을 볼 수 있었다.

#### 모델 3 : ARMA(1,1)

```r
par(mfrow=c(1,2))
h=2:10
ACF3 <- c(1,0.66187,(0.5^(h-1))*0.66187)
r_2 <- 0.5*0.66187
PACF3 <- c(0.66187,(r_2-0.66187^2)/(1-0.66187^2))
z3_acf <- acf(z3,plot=F)
plot(z3_acf[0:10])
points(x=0:10,y=ACF3,col="red",lwd=2)
z3_pacf <- pacf(z3,plot=F)
plot(z3_pacf[0:10])
points(x=1:2,y=PACF3,col="red",lwd=2)
```

* 표본의 크기를 만 개로 늘렸을 때 3모형 모두 이론적인 형태와 매우 유사한 형태를 보였다,
* 이는 표본의 크기가 커짐에 따라 추정된 값이 실제 모수 값의 가까워 진다는 대수의 법칙과 유사하다.
