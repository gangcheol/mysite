# 0. 정상 ARAM 모형의 적합 절차

`1`. 정상성 판단 : 비정상시계열의 경우 변환, 분해법, 차분 등을 통해 정상시계열로 변환

`2`. 모형의 식별 : $ARMA(p.q)$ 차수 선택

`3`. 모형의 추정 : 우리가 설정한 모델의 대하여 모수를 추정하는 것이다.

`4`. 모형의 진단 : 화이트 노이즈라는 것에 대한 진단이 필요하다(잔차 분석). 화이트 노이즈 가정에 적합하지 않으면 다시 모형의 식별 단계로 돌아간다.

`5.`예측모형 : 최종적으로 채택

# 1. 모형 식별 (model identification)

`1`. 모형을 적합하기 위한 전처리 단계라고 생각하자.

`step .1` 정상화 : 분산 안정화변환(이분산성 제거) -> 분해법을 이용하거나 차분, 계절 차분 시행(SACF를 그려 확인)

* 과대차분을 하지 않도록 주의하여야 한다.

* 과대차분을 확인하기 위해선 차분된 시계열의 분산과 이전 단계의 분산을 비교 $Var(\nabla Z) > Var(Z)?$ 좌변이 더 클경우 과대차분의 가능성이 보인다.

* 샘플사이즈가 적을 경우 $SACF$의 도표를 보고 추세가 존재하는지 판단하기 힘들다.

* 추세나 계절성이 실제로 존재하는지 판단하기 위해 **Unit root test(단위 근 검정)** 결과를 이용하여,

example $$AR(1) \quad H_0 : \phi =1 \quad vs \quad H_1 :\phi \neq 1$$

    * 귀무가설을 기각시 정상성을 만족한다고 볼 수 있다.

 * 계절형 자료가 아닌 경우 대부분 차분의 차수는 0, 1, 2이면 충분하고, 계절형 자료의 경우는 계절의 주기와 같은 값의 차분이면 충분하다.

* AR(p)의 경우 

$$H_0 : \phi_{kk} =0 \,vs\, H_1 : \phi_{kk}\neq0 \to |\hat {\phi}_{kk}|>1.96\,/\,\sqrt n \approx  2 / \sqrt n\quad   ( k\geq p+1)$$

*  MA(q)의 경우

$$H_0 : \rho_{k} =0 \,vs\, H_1 : \rho_{k}\neq0 \to |\hat {\rho}_{k}|>1.96\,\times\,S.E(\hat {\rho}_k)\quad   ( k\geq p+1)$$

$$\begin{eqnarray*}Var\,(\hat {\rho}_k) = \frac 1 n (1+2\rho_1^2+\dots 2p_q^2) = \frac 1n \end{eqnarray*}$$

화이트 노이즈를 가정하기 때문에 $Var\,(\hat {\rho}_k)  =\frac 1n$이 성립한다. (그냥 그렇다고만)

* ARMA(p,q)의 경우 : 여러 가지 후보모형을 생각해서 다 적합시킨다.( 모형 선택 기준은 아래와 같다.)

`1.` : $AIC$(Akaike’s Information Criterion, Akaike)

$$ AIC = n\ln(\hat {\sigma}^2)+2(p+q)$$

`2.` : $SBC$ (Schwartz’s Bayesian Criterion, Schwartz (1978)) 

$$ SBC = n\ln(\hat {\sigma}^2) +(p+q)\ln(n)$$

* 즉, $AIC, SBC, FPE, BIC, CAT$ 등이 최소가 되도록 모형 선택

`step 3.`상수항 포함 여부

* 시계열을 차분한 경우, 상수항 $\delta$를 모형에 포함시킬지의 여부를 결정
    * 이 말은 $E(\nabla Z_t) = 0$으로 할지 말지에 대한 결정이다.
    * 즉 $\mu=0 $으로 할지 말지 결정하는 것이다.

* 차분된 시계열 $W_t = (1-B)^dZ_t$의 표본평균과 근사표준오차를 이용하여 $H_0: \delta = 0$의 유의성 검정을 시행

* 검정 통계량 : $\begin {eqnarray*}\frac {\overline W}{S_{\overline W}} \end{eqnarray*}$

* $S_{\overline W} : \overline {W}_t$의 표본 분산

* $\begin{eqnarray*}\overline W = \frac 1n \sum W_t, S_{\overline W}= \left [\frac {\hat {\gamma}_0}{n}(1+2\hat{\rho}_1+\dots 2\hat{\rho}_k)\right]^{1/2} \end{eqnarray*}$ 단, $k$는 유의한 $SACF$의 개수

* $\hat {\gamma}_0 :$ $W_t$ 의 표본분산 

* 위에는 실제로 계산하는게 아니니 깊게 생각하지 말자...

# 2. 모수의 추정 (Parameter Esitmation)

* 모형을 식별 후 식별한 모형의 모수, 즉 특성을 파악하는 단계이다.

* 적률추정법, 최소제곱법(CLSE, ULSE), 최대가능도함수를 이용하여 추정

* 적률추정법의 경우 최종적인 추정방법으로는 추정할수는 없고, 다른 추정방법의 초기치로 사용됨

## 2-1. 적률추정법 AR(1)

$$Z_t = \phi(Z_{t-1}-\mu)+\varepsilon_t \quad \varepsilon_t \sim WN(0,\sigma^2)$$

* $\mu = E(Z_t) \to \hat {\mu}_{MME} = \overline Z$

* $\begin {eqnarray*}Cov(Z_t,Z_{t-1}) = \gamma_1 = \phi\gamma_0 \to \hat {\phi}_{MME} = \frac {\hat {\gamma}_1}{\hat {\gamma}_0} \end{eqnarray*}$

* $Var(Z_t) = \phi^2Var(Z_{t-1}) +\sigma^2 \to \gamma_0 = \phi^2 \gamma_0 + \sigma^2$

$\therefore \quad \hat {\sigma}^2_{MME} = \hat {\gamma}_0 (1-\hat{\phi}^2)$

## 2-2. 적률추정법 MA(1)

$$Z_t -\mu = \varepsilon_t - \theta \varepsilon_{t-1}$$

* $\hat {\mu}_{MME} = \overline Z$

* $\begin{eqnarray*} \hat {\theta}_{MME} = \frac {-1 \pm \sqrt{1-4\hat {\rho}_1^2}}{2\hat{\rho}_1}\end{eqnarray*}$

  $if \quad \hat {\rho}_1 = \pm \,0.5 \to \hat {\theta} = \pm 1$ 이므로 가역성을 만족하지 않아 추정량으로 쓸 수 없음

  $eles \,if \quad |\hat {\rho}_1| > \pm \,0.5 \to \hat {\theta}$ 는 허수로 존재하지 않음

  $eles \,if \quad |\hat {\rho}_1| < \pm \,0.5 \to |\hat {\theta}|<   1$ 인 경우, 즉 가역성을 만족하는 경우를 찾아서 사용

* $\begin {eqnarray*}  \hat {\sigma}^2 = \frac {\hat {\gamma_0}}{1+\theta^2}\end{eqnarray*}$

조건부, 비조건부 최소제곱법과 MLE 추정법은 강의노트 참고

# 3. 잔차분석 (Residual Analysis)

* 잔차를 이용하여 $\varepsilon_t \sim WN(0,\sigma^2)$ 인지를 검정

* 잔차를 이용한 **WN TEST( Portmanteau test )** <br>
    * $H_0: \rho(\varepsilon_1) = \dots \rho_k(\varepsilon_t) = 0$
    * 위를 검정하며 "**box-pierce test", "Ljung-Box test"** 가 있다,
    * 후자의 경우 샘플사이즈가 작은 경우 검정통계량 값을 크게하기 위하여 사용한다.

* "**Jarque and Bera**"의 정규성검정

$$\begin {eqnarray*} JB = \frac n6 \left( \frac 1n\sum \frac {\varepsilon_t^3}{\hat \sigma^2} \right )+\left( \frac 1{24}\sum \frac {\varepsilon_t^4}{\hat \sigma^2}-3 \right ) \approx \chi^2(2)\end{eqnarray*}$$

* 좌항을 왜도, 우항을 첨도로 하여 왜도와 첨도를 가지고 정규성 검정을 시행하는 것이다.

# 4. 과대적합 진단 (Overfitting diagostics)

* 모형에 모수를 추가해나가며 과댁적합 진단을 시행

* 다음과 같은 경우 기존의 잠정모형을 새로운 모형으로 대체한다.

`1` 잠정모형에 추가된 모수가 유의하다고 판정되는 경우

`2` 잠정모형에서의 모수 추정값이 과대적합시킨 후 얻어지는 모수 추정값과 큰 차이가 있는 경우

`3` 과대적합된 모형에서 잔차들의 분산이 잠정모형에서의 잔차들의 분산보다 작아진 경우

# 5. 참고 : 단위근 검정

$\Phi(B) = (1-\phi_1 B -\dots - \phi_pB)$

$$H_0: \phi=1\quad vs \quad H_1 : |\phi| <1$$

* 차분이 필요한 지 아닌지 검정한다.

* $1-B$에서 $B=1$ 인 경우 단위근이라고 하는데 이걸 가지는 경우 비정상 시계열이므로 $ARIMA$ 모형을 적합해야함.

* 단위근이 있는 경우가 $\phi =1 $인 경우이다.

* 차분을 통해 정상 확률 과정으로 변환 가능

* R 에서 "**Dickey -Fuller**"로 단위근 검정을 시행 (ex AR(1))
    * `case 1` : 적합할 모형에 절편(상수항)이 없는 경우 (nc) 평균이 0인 경우
    * `case 2` : 적합할 모형에 절편이 존재하는 경우 (c) 평균이 0이 아닌 경우 
    * `case 3` : 적합할 모형에 절편과 추세($\beta_t$)가 존재하는 경우 (ct) 추세가 있어보이는 경우

## 예제 8.6 가스로 자료 분석

* 가스로 공급비율과 이에 따른 이산화탄소의 농도를 나타낸다. 여기서는 가스의 공급비율에 적합한 모형을 적합해보자.

### 데이터 로드 및 시도표 확인

```r
options(repr.plot.res=200,repr.plot.width=10,repr.plot.height=5)
```

```r
z <- scan("gas.txt", what=list(0,0))
```

```r
dt <- tibble(t= 1:length(z[[1]]),
            rate = z[[1]],
            co2 = z[[2]])
```

```r
p3 <- ggplot(dt, aes(t, rate)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+ ggtitle('Time serise plot of the input Gas feed rate')+
  theme_bw()

p1 <- ggAcf(dt$rate) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$rate) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())



grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

* 시계열그림을 살펴본 결과 추세는 없는 것처럼 보이며 ACF와 PACF값을 보아도 딱히 문제될 것이 없어 보인다.

* 단위근 검정을 통해 실제로 정상성 여부를 판단한다.

* 또한 시도표를 살펴본 결과 평균을 0, 즉 상수항이 없다고 판단할 수 있다.

### 단위근 검정

```r
options(warn=-1)
```

`1` 

```r
adfTest(dt$rate, lags = 0, type = "nc") 
```

```r
adfTest(dt$rate, lags = 0, type = "nc")
adfTest(dt$rate, lags = 1, type = "nc")
adfTest(dt$rate, lags = 2, type = "nc")
```

* 각 시차에서의 $p-value$값을 살펴본 결과 단위근이 없다. 즉, 확률적 추세가 없으므로 차분을 할 필요가 없다고 판단된다.

* 즉 우리가 불러온 데이터가 ACF 지수적으로 감소하고, PACF가 시차 3까지 유의 하므로 AR(3) 모형 정도가 된다고 잠정적으로 식별한다.

### 모형적합

* AR(3) 모형 적합

```r
fit1 <- arima(dt$rate, order=c(3,0,0))
```

```r
summary(fit1)
```

* 추정된 계수에 비해 s.e 가 너무 크다면 유의하지 않다고 summary값으로 간략히 확인할 수 있다.

* 만약 intercept가 작은데 s.e 가 크다면 추정된 계수가 유의하지 않다고 볼 수 있다.

* 모수 추정의 디폴트 방법은 조건부최소제곱법(css)와 최대가능도함수를 이용한다.

```r
coeftest(fit1)
```

* 또한 추정된 $\phi_1,\phi_2.\phi_3$ 이 유의미함을 알 수 있다.
* 또한 절편의 경우 유의미하지 않음을 확인할 수 있었다.

```r
fit2 <- arima(dt$rate, order=c(3,0,0), include.mean = F)
summary(fit2)
coeftest(fit2)
```

* `include.mean = F`, 상수항 ,즉 절편을 빼는지에 대한 결정 여부이다.

### ACF 유의성 검정

```r
LjungBox(dt$rate, lags =seq(6,24,6)) ## 포트맨토 검정
```

* 전부 유의확률이 0이므로 해당 시차에 $\rho(\varepsilon_t)\neq0$ 이다.

* 여기까지 잠정모형으로 fit2을 선택

### 잔차검정

```r
dt["resid"] = as.numeric(resid(fit2)) ## 잔차 생성
```

```r
p3 <- ggplot(dt, aes(t, resid)) + 
  geom_line(col='steelblue') +
  geom_hline(yintercept = 0, col='grey', lty=2)+
  xlab("")+ylab('')+ ggtitle('Time serise plot of Residual')+
  theme_bw()

p1 <- ggAcf(dt$resid) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$resid) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

* 잔차에 대한 시도표와 ACF, PACF 그림을 그린 결과이다.

* 살펴본 결과 화이트노이즈 가정을 위배하는 것처럼 보이진 않으나 중간중간 이상치가 보인다.

#### 잔차의 정규성 검정

```r
shapiro.test(dt$resid)  ##H0 : normal distribution
```

* 정규분포를 기각한 결과가 나오는데 이상치로 판단된다.

```r
qqnorm(dt$resid)
qqline(dt$resid, col = 2) 
```

* 양쪽 꼬리부분에서 정규성에 위반된 형태를 띈다.

```r
LjungBox(fit2, lags=seq(6,24,6)) ##포트만트 검정
```

* ACF가 0이라는 가정을 기각하는 경우가 많아 화이트 노이즈가 의심된다.

* 정규성과 ACF가 0이라는 가정이 전부 기각되었다. 이상치 때문이라는 의심이 판단된다.

* 과대적합(차수를 늘려가며)을 통해 다른 모형을 고려할 필요가 있어보인다.

* 다양한 모형을 적합해가며 살펴본 결과 결과가 나아지지 않아 AR(3)을 최종모형으로 선정하였다.

* 즉, 과대적합을 굳이 할 필요가 없어 보인다.

```r
p3 <- ggplot(dt, aes(t, z)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+ ggtitle('Time serise plot')+
  theme_bw()

p1 <- ggAcf(dt$z) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$z) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

* 단위근이 있다는 것은 비정상 시계열이고 차분할 필요가 있다는 뜻이므로 차분을 시행

### 차분

```r
dt["diff_z"] = c(0,diff(dt$z))
```

```r
p3 <- ggplot(dt, aes(t, diff_z)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+ ggtitle('Time serise plot')+
  theme_bw()

p1 <- ggAcf(dt$diff_z) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$diff_z) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

* 차분한 결과 ACF와 PACF가 화이트 노이즈 처럼 나왔다.

```r
LjungBox(dt$diff_z, lags=seq(6,24,6))
```

```r
t.test(dt$diff_z)
```

* 평균이 0이라고 해도 무방하다.

* 차분한 결과 화이트노이즈라고 봐도 무방하다.

```r
fit <- arima(dt$z,order=c(0,1,0))
summary(fit)
```

```r
coeftest(fit) 
```

* 에러가 나는 이유는 화이트 노이즈이기 때문에 모수를 추정할 필요가 없기 때문이다.

```r
dt["resid"]  <- as.numeric(resid(fit))
```

```r
p3 <- ggplot(dt, aes(t, resid)) + 
  geom_line(col='steelblue') +
  geom_hline(yintercept = 0, col='grey', lty=2)+
  xlab("")+ylab('')+ ggtitle('Time serise plot of Residual')+
  theme_bw()

p1 <- ggAcf(dt$resid) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$resid) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

```r
shapiro.test(dt$resid)  ##H0 : normal distribution
jarque.bera.test(dt$resid)  ##JB test H0: normal
```

```r
qqnorm(dt$resid)
qqline(dt$resid, col = 2)
```

```r
LjungBox(fit, lags=seq(6,24,3))
```

* 화이트노이즈인 것은 문제가 없으나 정규성 문제가 의심된다.

* 그러나 시계열에서는 정규성을 가지고 문제를 다루지 않기 때문에 딱히 문제가 될 것은 없다.

```r
adfTest(dt$z, lags = 0, type = "ct")
adfTest(dt$z, lags = 1, type = "ct")
adfTest(dt$z, lags = 2, type = "ct") 
```

### 선형모형 적합

* 결정적 추세가 있어보여 선형모형을 적합시켜봄

```r
fit1 <- lm(z~t, data=dt)
summary(fit1)
coefficients(fit1)
```

```r
dt[, resid_lm := residuals(fit1)]
```

```r
p3 <- ggplot(dt, aes(t, resid_lm)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+ ggtitle('Time serise plot')+
  theme_bw()

p1 <- ggAcf(dt$resid_lm) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$resid_lm) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())
```

```r
grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

* ACF의 경우 여전히 천천히 감소한다.

```r
LjungBox(dt$resid_lm, lags=seq(6,24,6)) 
```

```r
adfTest(dt$resid_lm, lags = 0, type = "nc")
adfTest(dt$resid_lm, lags = 1, type = "nc")
adfTest(dt$resid_lm, lags = 2, type = "nc") 
```

```r
jarque.bera.test(dt$resid_lm)
```

* 단위근은 없다고 판단된다. 그러나 ACF를 그려본 결과 차분을 할 필요성이 있다고 판단된다. 

* 다시 말하면 우리는 결정적 추세가 있어보여 추세모형을 적합시켰다.

* 거기서 얻어진 추세성분을 제거하여 잔차가 정상시계열이라고 생각을 했으나 잔차의 ACF값을 보니 차분할 필요가 있다고 판단이 되며 PACF값을 살펴본 결과 시차 1이후에 유의하지 않으므로 잔차에 $AR(1)$ 모형을 적합

### 잔차에 AR(1) 모형을 적합 후 차분

```r
fit_resid <- arima(dt$resid_lm, order=c(1,0,0), include.mean = F)
fit_resid
coeftest(fit_resid)
```

```r
dt[, diff_z := c(0, diff(z))]

p3 <- ggplot(dt, aes(t, diff_z)) + 
  geom_line(col='steelblue') +
  xlab("")+ylab('')+ ggtitle('Time serise plot')+
  theme_bw()

p1 <- ggAcf(dt$diff_z) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$diff_z) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

```r
LjungBox(dt$diff_z, lags=seq(6,24,6))
```

* 차분한 결과 whitenoise처럼 보인다.

```r
t.test(dt$diff_z)
```

* 그러나 평균은 0이 아니다. (0이 아니라고 정상시계열이 아니라는 것은 아님 0은 단지 화이트노이즈의 여부 중 하나)

* 차분을 한번 하였을 때 정상기계열이 되었으므로 $ARIMA(0,1,0)$을 적합

* $ARMIA (p,1,0) \sim AR(p+1)$ 임을 기억하자 (챕터 7)

### ARIMA (0,1,0)

```r
fit <- arima(dt$z,order=c(0,1,0),
             include.mean = TRUE)
summary(fit)
```

#### 잔차분석

```r
dt[, resid := as.numeric(resid(fit))]

p3 <- ggplot(dt, aes(t, resid)) + 
  geom_line(col='steelblue') +
  geom_hline(yintercept = 0, col='grey', lty=2)+
  xlab("")+ylab('')+ ggtitle('Time serise plot of Residual')+
  theme_bw()

p1 <- ggAcf(dt$resid) + 
  theme_bw() + ylim(-1,1) +
  theme(plot.title = element_blank())

p2 <- ggPacf(dt$resid) + 
  theme_bw() +ylim(-1,1) +
  theme(plot.title = element_blank())

grid.arrange(p3, p1, p2, nrow = 2,
             layout_matrix = rbind(c(1,1),
                                   c(2,3)))
```

* 앞서 차분한 결과와 동일하게 나옴

#### 잔차분석

```r
shapiro.test(dt$resid)  ##H0 : normal distribution
jarque.bera.test(dt$resid)  ##JB test H0: normal
```

* 근데 왠만하면 선형추세모형보다 ARIMA 모형을 쓰는 것이 복잡하지 않고 간편하다.
